```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("categorical_retail_data.csv")
df.head()
```

```python
df.select_dtypes(include='object').columns

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['store_enc'] = le.fit_transform(df['store'])
df['region_enc'] = le.fit_transform(df['region'])
```

```python
df = pd.get_dummies(
    df,
    columns=['category', 'store_type'],
    drop_first=True
)
```

```python
brand_freq = df['brand'].value_counts()
df['brand_freq'] = df['brand'].map(brand_freq)
```

```python
product_target_mean = df.groupby('product')['sales'].mean()
df['product_target_enc'] = df['product'].map(product_target_mean)
```

```python
print("Dataset shape:", df.shape)
```

```python
df.drop(
    columns=['store','region','brand','product'],
    inplace=True
)
```

```python
df.to_csv("engineered_retail_data.csv", index=False)

print("✅ engineered_retail_data.csv created successfully")
df.head()
```
